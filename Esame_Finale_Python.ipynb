```python
# Importazione delle librerie python e caricamento dati Covid tramite file .xlsx

import numpy as np
import pandas as pd
df = pd.read_excel("C:\\Users\\leo\\Desktop\\owid-covid-data.xlsx")
df
```

```python
# Verifico le dimensioni del dataset e i relativi metadati 

print("Dimensioni e riepilogo del Dataframe:\n",df.shape)
df.info()
df.head()
```

```python
# Rimuovo i valori che non appartengono ad alcun continente

df_nuovo = df.dropna(subset="continent")
df_nuovo
```

```python
# Numero di casi totali avvenuti nello stesso continente

numero_casi_continete = df_nuovo.groupby("continent")["new_cases"].agg(casi_totali=(sum))
print(numero_casi_continete.astype(int))
```

```python
# Casi totali nel mondo calcolati anche sulle locazioni senza indicazione di continente

totale_casi_mondo = df["new_cases"].sum()
print(totale_casi_mondo)
```

```python
dati_dei_continenti = df.groupby("continent")["new_cases"].agg(sum)
print(dati_dei_continenti.astype(int))
```

```python
# Date due variabili contenenti i nomi di continenti confrontare valore massimo, media, e percentuale rispetto ai casi totali


def comparazionecasi(continente1, continente2):
    
    totale_casi_mondo = df["new_cases"].sum()

    casicontinente1 = dati_dei_continenti[continente1]
    casicontinente2 = dati_dei_continenti[continente2]

    mediacontinente1 = df[df["continent"] == continente1]["new_cases"].mean()
    mediacontinente2 = df[df["continent"] == continente2]["new_cases"].mean()

    massimocontinente1 = df[df["continent"] == continente1]["new_cases"].max()
    massimocontinente2 = df[df["continent"] == continente2]["new_cases"].max()

    percentualecontinente1 = (casicontinente1 / totale_casi_mondo) * 100
    percentualecontinente2 = (casicontinente2 / totale_casi_mondo) * 100

    comparazione = {continente1: {"massimo": massimocontinente1, "media": mediacontinente1, "percentuale": percentualecontinente1},
        continente2: {"massimo": massimocontinente2, "media": mediacontinente1, "percentuale": percentualecontinente1}}

    return comparazione

continente1 = "Oceania"
continente2 = "Europe"
comparazione = comparazionecasi(continente1, continente2)
dataframecomparazione = pd.DataFrame(comparazione)
print(dataframecomparazione.round(2))
```

```python
# Andamento Casi Covid-19 in Italia anno 2022

df["date"] = pd.to_datetime(df["date"])
dati_italia2022 = df[(df["location"] == "Italy") & (df["date"].dt.year == 2022)]

import matplotlib.pyplot as plt

plt.figure(figsize=(7, 3))
plt.plot(dati_italia2022["date"], dati_italia2022["total_cases"])
plt.title("Casi Covid-19 in Italia anno 2022")
plt.xlabel("Data")
plt.ylabel("Casi Totali")
plt.grid(True)
plt.show()
```

```python
# Andamento nuovi casi Covid-19 in Italia anno 2022

newdati_italia2022 = df[(df["location"] == "Italy") & (df["date"].dt.year == 2022)]

plt.figure(figsize=(7, 3))
plt.plot(newdati_italia2022["date"], dati_italia2022["new_cases"])
plt.title("Andamento nuovi casi Covid-19 in Italia anno 2022")
plt.xlabel("Data")
plt.ylabel("Nuovi Casi")
plt.grid(True)
plt.show()
```

```python
# Andamento Comulativo nuovi casi Covid-19 in Italia anno 2022

newdati_italia2022 = df[(df["location"] == "Italy") & (df["date"].dt.year == 2022)]
newdati_italia2022["somma_comulativa"] = newdati_italia2022["new_cases"].cumsum()

plt.figure(figsize=(7, 3))
plt.plot(newdati_italia2022['date'], newdati_italia2022["somma_comulativa"])
plt.title("Andamento Comulativo nuovi casi Covid-19 in Italia anno 2022")
plt.xlabel("Data")
plt.ylabel("Somma Nuovi Casi")
plt.grid(True)
plt.show()

print("Guardando attenamente il primo grafico e il terzo grafico notiamo una similarità di andamento.\nProbabilmente nel periodo del anno 2022 le guarigioni andavano molto a rillento rispetto ai nuovi casi")
```

```python
# Filtro i dati per Italia, Germania e Francia da maggio 2022 ad aprile 2023


df["date"] = pd.to_datetime(df["date"])

data_inizio = "2022-05-01"
data_fine = "2023-04-30"
stati = ["Italy", "Germany", "France"]
nuovidati = df[(df["location"].isin(stati)) & 
                     (df["date"] >= data_inizio) & 
                     (df["date"] <= data_fine)]

nuovidati = nuovidati.dropna(subset=["icu_patients"])
nuovidati = nuovidati[["date","location", "icu_patients"]]
nuovidati

# Creo boxplot per vedere la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva
plt.figure(figsize=(7, 3))
nuovidati.boxplot(column="icu_patients", by="location",grid=False)
plt.title("Numero di pazienti in terapia intensiva da Maggio 2022 ad Aprile 2023")
plt.xlabel("Stati")
plt.ylabel("Numero di pazienti in terapia intensiva")
plt.show()
```

```python
# Somma Numero di Pazienti Ospitalizzati per Italia, Germania, Spagna e Francia
data_inizio = "2023-01-01"
data_fine = "2023-12-31"
stati = ["Italy", "Germany", "Spain", "France"]
nuovidati2 = df[(df["location"].isin(stati)) & 
                     (df["date"] >= data_inizio) & 
                     (df["date"] <= data_fine)]

nuovidati2 = nuovidati2[["location", "hosp_patients"]]
nuovidati2 = nuovidati2.dropna(subset=["hosp_patients"])
sommaospedalizzati = nuovidati2.groupby("location")["hosp_patients"].sum()

plt.figure(figsize=(10, 3))
sommaospedalizzati.plot(kind="bar", color=["red", "orange", "blue", "red"])
plt.title("Somma dei Pazienti Ospitalizzati nel 2023")
plt.xlabel("Paese")
plt.ylabel("Numero di Pazienti Ospitalizzati")
plt.grid(True)
plt.show()

# La colonna dei paziente ospedalizzati della Germania risulta prima di dati

df_mask = nuovidati2["location"] == "Germany"
filtered_df = nuovidati2[df_mask]
print(filtered_df)
```
