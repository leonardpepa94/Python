```python
# Importazione delle librerie python e caricamento dati Covid tramite file .xlsx

import numpy as np
import pandas as pd
df = pd.read_excel("C:\\Users\\leo\\Desktop\\owid-covid-data.xlsx")
df
```

```python
# Verifico le dimensioni del dataset e i relativi metadati 

print("Dimensioni e riepilogo del Dataframe:\n",df.shape)
df.info()
df.head()
```

```python
# Rimuovo i valori che non appartengono ad alcun continente

df_nuovo = df.dropna(subset="continent")
df_nuovo
```

```python
# Numero di casi totali avvenuti nello stesso continente

numero_casi_continete = df_nuovo.groupby("continent")["new_cases"].agg(casi_totali=(sum))
print(numero_casi_continete.astype(int))
```

```python
# Casi totali nel mondo calcolati anche sulle locazioni senza indicazione di continente

totale_casi_mondo = df["new_cases"].sum()
print(totale_casi_mondo)
```
